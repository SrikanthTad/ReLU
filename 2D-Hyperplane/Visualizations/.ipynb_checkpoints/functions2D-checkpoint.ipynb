```python
#Return points associated with line x,y seperately
def Generate_Line(fn, x):
    
    '''
    x is number of points wanted
    fn = [c,a,b] array input
    line: c + a*x + b*y = 0
    
    '''  
    assert not (fn[1]==0 and fn[2]==0), "coefficients for both x and y are zero"
    
    #x = np.linspace(-5,5,1000)
    
    if fn[2]!=0:
        
        y = -fn[1]/fn[2]*x-fn[0]/fn[2]
        
    else:
        
        y = x
        x = np.full_like(y, -fn[0]/fn[1])
        

    return x,y

#ReLU function
def relu(x):
    return x * (x > 0)

#single list ReLU
def applyRelu(ls):
    ls2 =[]
    for i in ls:
        ls2.append(relu(i))
    return ls2
```

```python
#appends two list to make points [x,y]
def makeList(ls1,ls2):
    ls = []
    for i in range(len(ls1)):
        ls.append([ls1[i],ls2[i]])
    return ls

#adds [1,x,y]
def makeList1(ls1,ls2):
    ls = []
    for i in range(len(ls1)):
        ls.append([1,ls1[i],ls2[i]])
    return ls

#gives 2 list of x=[] and y=[] points
def seperatePoints(ls):
    '''ls = [1,x,y]'''
    x = []
    y = []
    for i in ls:
        x.append(i[1])
        y.append(i[2])
    return x,y

#Applies forward weight matrix
def applyWeights_on_original(weights_matrix, data_matrix):
    '''make data_matrix values into x = [1,x,y]'''
    ls0 = []
    for x in data_matrix:
        ls0.append([1,x[0],x[1]])
    ls1 = []
    for i in ls0:
        ls1.append(np.matmul(weights_matrix,i))
    return ls1

def applyWeights(weights_matrix,data_matrix):
    '''data matrix already has values of x = [1,x,y]'''
    ls0=[]
    for i in data_matrix:
        ls0.append(np.matmul(weights_matrix,i))
    return ls0

#Applies inverse weight matrix
def applyWeightsInverse(weights_matrix, data_matrix):
    '''data matrix already has values of x = [1,x,y]'''
    weights_inverse = np.linalg.inv(weights_matrix)
    ls1 = []
    for x in data_matrix:
        ls1.append(np.matmul(weights_inverse,x))
    return ls1

def matrix_from_weights(Matrix, bias):
    W = np.hstack((Matrix, bias[:, np.newaxis]))
    W = np.vstack((W, np.array([0,0,1])))
    return W

def intersection(L1, L2):
    D  = L1[0] * L2[1] - L1[1] * L2[0]
    Dx = L1[2] * L2[1] - L1[1] * L2[2]
    Dy = L1[0] * L2[2] - L1[2] * L2[0]
    if D != 0:
        x = Dx / D
        y = Dy / D
        return x,y
    else:
        return False

def line(p1, p2):
    A = (p1[1] - p2[1])
    B = (p2[0] - p1[0])
    C = (p1[0]*p2[1] - p2[0]*p1[1])
    return A, B, -C

def makeArrows(ls1,ls2):
    for i in range(len(ls1)):
        plt.arrow(ls1[0],ls1[1], ls2[0],ls2[1] , head_width=0.5, head_length=0.5)
```
